Task 1 — A* Search (8 Puzzle)

```python
import heapq

goal = [1,2,3,
        8,0,4,
        7,6,5]

start = [2,8,3,
         1,6,4,
         7,0,5]


def manhattan(state):
    distance = 0
    for i in range(9):
        if state[i] != 0:
            goal_index = goal.index(state[i])
            x1,y1 = divmod(i,3)
            x2,y2 = divmod(goal_index,3)
            distance += abs(x1-x2) + abs(y1-y2)
    return distance


def get_neighbors(state):
    neighbors = []
    blank = state.index(0)

    moves = {
        "up": -3,
        "down": 3,
        "left": -1,
        "right": 1
    }

    for move, step in moves.items():

        new_blank = blank + step

        if move == "up" and blank < 3:
            continue
        if move == "down" and blank > 5:
            continue
        if move == "left" and blank % 3 == 0:
            continue
        if move == "right" and blank % 3 == 2:
            continue

        new_state = state.copy()
        new_state[blank], new_state[new_blank] = new_state[new_blank], new_state[blank]

        neighbors.append(new_state)

    return neighbors


def astar(start):

    open_list = []
    heapq.heappush(open_list,(manhattan(start),0,start))

    visited = set()

    while open_list:

        f,g,state = heapq.heappop(open_list)

        if state == goal:
            print("Goal Reached:",state)
            return

        visited.add(tuple(state))

        for neighbor in get_neighbors(state):

            if tuple(neighbor) not in visited:

                heapq.heappush(open_list,(g+1+manhattan(neighbor),g+1,neighbor))


astar(start)
```

Task 2 — A* Example (Shortest Path in Graph)

```python
import heapq

graph = {
'A':[('B',1),('C',3)],
'B':[('D',3),('E',1)],
'C':[('F',5)],
'D':[],
'E':[('F',2)],
'F':[]
}

heuristic = {
'A':5,
'B':4,
'C':4,
'D':2,
'E':1,
'F':0
}

def astar(start,goal):

    open_list = []
    heapq.heappush(open_list,(0,start))

    cost = {start:0}

    while open_list:

        f,node = heapq.heappop(open_list)

        if node == goal:
            print("Goal reached:",node)
            return

        for neighbor,weight in graph[node]:

            g = cost[node] + weight

            if neighbor not in cost or g < cost[neighbor]:

                cost[neighbor] = g
                f = g + heuristic[neighbor]

                heapq.heappush(open_list,(f,neighbor))


astar('A','F')
```
