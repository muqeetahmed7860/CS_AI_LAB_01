We are implmenting List

Create an initial list

```python
my_list = [1, 2, 3, 4, 5]
print(my_list)
```

1. append() Adds an element at the end of the list

```python
my_list.append(6)
print("After append:", my_list)
```

 2.Clear() Removes all the elements from the list

```python
my_list.clear()
print("After clear:", my_list)
```

Reinitialize the list

```python
my_list = [1, 2, 3, 4, 5]
```


3. copy() Returns a copy of the list

```python
copied_list = my_list.copy()
print("Copied list:", copied_list)
```

 4. Count()  Returns the number of occurrences of the specified value

```python
count_3 = my_list.count(3)
print("Count of 3:", count_3)
```

5. extend()  Adds the elements of another list to the end of the current list

```python
my_list.extend([6, 7])
print("After extend:", my_list)
```

6. index()  Returns the index of the first occurrence of the specified value

```python
index_of_4 = my_list.index(4)
print("Index of 4:", index_of_4)
```

7. insert() Adds an element at the specified position

```python
my_list.insert(2, 10)  # Insert 10 at index 2
print("After insert:", my_list)
```

8. pop() Removes the element at the specified position (last if no index specified)

```python
popped_item = my_list.pop()
print("Popped item:", popped_item)
print("After pop:", my_list)
```

9. remove() Removes the first occurrence of the specified value

```python
my_list.remove(10)
print("After remove:", my_list)
```

10. Reverse()  Reverses the order of the list

```python
my_list.reverse()
print("After reverse:", my_list)
```

11. sort()  Sorts the list

```python
my_list.sort()
print("After sort:", my_list)
```
