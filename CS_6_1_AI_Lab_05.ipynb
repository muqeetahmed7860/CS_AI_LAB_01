(Code Cell – Graph Definition)

```python
graph = {
    0: [1, 2],
    1: [0, 2, 3],
    2: [0, 1, 4],
    3: [1, 4],
    4: [2, 3]
}

print("Graph Representation:")
for node in graph:
    print("Node", node, "->", graph[node])
```

(Code Cell – Depth First Search)

```python
visited_dfs = set()

def dfs(node):
    if node not in visited_dfs:
        print("Visited:", node)
        print("Adjacent Nodes of", node, ":", graph[node])
        visited_dfs.add(node)

        for neighbor in graph[node]:
            dfs(neighbor)

print("\nDepth First Search Traversal:")
dfs(0)
```

(Code Cell – Breadth First Search)

```python
from collections import deque

def bfs(start):
    visited_bfs = set()
    queue = deque([start])
    visited_bfs.add(start)

    while queue:
        node = queue.popleft()
        print("Visited:", node)
        print("Adjacent Nodes of", node, ":", graph[node])

        for neighbor in graph[node]:
            if neighbor not in visited_bfs:
                visited_bfs.add(neighbor)
                queue.append(neighbor)

print("\nBreadth First Search Traversal:")
bfs(0)
```
