```python
def is_prime(n):
    if n == 0 or n == 1:
        return 0
    else:
        for i in range(2, n):
            if n % i == 0:
                return 0
    return 1
```

```python
x = int (input("Enter a number"))
if is_prime(x)==0:
    print ("number is not prime")
else :
    print("number is prime")
```

```python
count = 1
for i in range(100):
    if is_prime(i) == 1:
        print(count, "\t", i)
        if count >= 20:
            break
        count += 1
```
